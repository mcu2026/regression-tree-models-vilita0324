| 模型 | 訓練集 R² | 測試集 R² | 差距 (Train-Test) |
|------|-----------|-----------|------------------|
| Linear Regression   | 0.7652 | 0.5701 | 0.1951 |
| CART Regression Tree | 0.9474 | 0.6432 | 0.3042 |

### 分析

1. **Linear Regression（線性迴歸）**：
   - 訓練集 R² = 0.7652，測試集 R² = 0.5701
   - 差距較小（0.1951），泛化能力相對較穩定
   - 但整體 R² 偏低，代表線性模型對這個資料集的擬合能力有限

2. **CART Regression Tree（決策樹迴歸）**：
   - 訓練集 R² = 0.9474，測試集 R² = 0.6432
   - 差距較大（0.3042），有明顯overfitting的現象
   - 雖然測試集 R² 優於 Linear Regression，但模型複雜度較高

### 比較結論

- 測試集效能：CART（0.6432）> Linear Regression（0.5701）
- CART 預測能力較強，但有嚴重的overfitting
- 若要改善 CART 的過擬合，可設定 `max_depth` 或 `min_samples_leaf` ，增加模型的深度
